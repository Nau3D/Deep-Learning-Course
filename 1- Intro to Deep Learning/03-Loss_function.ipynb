## Loss Function

The loss function quantifies how bad a model is performing, the larger the value the worse the model.

It is a measure of the discrepancy between the model's output and the groud truth.

### Mean Squared Error

$ e = \large\sum_{i}(gt_i-pred_i)^2$

```python
import math
from matplotlib import pyplot as plt
import numpy as np
```

```python
def squared_error(gt, predicted):

    mean_square_error = (gt - predicted)**2.0
    return mean_square_error
```

```python
y = [squared_error(1, k/100) for k in range(100)]

x = [k/100 for k in range(100)]
plt.plot(x,y)
plt.show()
```

### Cross Entropy or Log loss

$ e =  - \large \sum_i gt_i \times log(pred_i)$

```python
def categorical_cross_entropy(gt, predicted):

    return  - gt * math.log(1e-8 + predicted)
```

```python
z = [categorical_cross_entropy(1, k/100) for k in range(100)]

plt.plot(x,z)
plt.show()
```

Cross Entropy has higher gradients pushing for large steps when the predicted value is further away ground truth.
